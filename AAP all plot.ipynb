```python
%cd /content/drive/MyDrive/Colab Notebooks/AA/github_main/AAP


import numpy as np
from scipy import optimize
import numpy
import time
import matplotlib.pyplot as plt
import sys

import Util.Logistic as Logistic
from Algorithm.Solver_NMF import Solver




class Demo:
    def __init__(self, maxiter, repeat, m_, solver):
        self.maxiter = maxiter
        self.repeat = repeat
        self.m_ = m_
        self.solver = solver

    #fit is a function to compute the global minimizer
    def fit(self, A, k):

        self.A = A
        self.k = k
        self.m,self.n = A.shape


    def testConvergence(self, isSearch=False):
        errMat = numpy.zeros((self.repeat*4, self.maxiter+1))
        picards=[]
        AAs=[]
        AAPs=[]
        resAAs=[]
        cur = 0
        for r in range(self.repeat):
            print(str(r) + '-th repeat  ')

            m_ = self.m_
            #np.random.seed(42)
            W0 = np.random.rand(self.m, self.k)
            #H0 = np.random.rand(self.k, self.n)


            print('+++++++   AAP   +++++++')
            solver = Solver(local_epochs = m_, eta0=.1, C=0, algo_='AAP')
            solver.fit(self.A, self.k, W0=W0, NMF_=3)
            start_time = time.time()
            err = solver.train( maxIter=self.maxiter,warmup=False, damping=1)
            print(cur, 'time used:   ' ,time.time() - start_time)
            print('error', err)
            print()
            l = min(len(err), self.maxiter+1)
            errMat[cur, 0:l] = err[0:l]
            AAPs.append(solver.funcValsPerPicard)
            cur = cur+1
            del solver


            print('+++++++   AA   +++++++')
            solver = Solver(local_epochs = m_, eta0=.1, C=0, algo_='AA')
            solver.fit(self.A, self.k, W0=W0, NMF_=3)
            start_time = time.time()
            err = solver.train( maxIter=self.maxiter,warmup=False, damping=1)
            print(cur, 'time used:   ' ,time.time() - start_time)
            print('error', err)
            print()
            l = min(len(err), self.maxiter+1)
            errMat[cur, 0:l] = err[0:l]
            AAs.append(solver.funcValsPerPicard)
            cur = cur+1
            del solver


            print('+++++++   resAA   +++++++')
            solver = Solver(local_epochs = m_, eta0=.1, C=0, algo_='resAA')
            solver.fit(self.A, self.k, W0=W0, NMF_=3)
            start_time = time.time()
            err = solver.train( maxIter=self.maxiter,warmup=False, damping=1)
            print(cur, 'time used:   ' ,time.time() - start_time)
            print('error', err)
            print()
            l = min(len(err), self.maxiter+1)
            errMat[cur, 0:l] = err[0:l]
            resAAs.append(solver.funcValsPerPicard)
            cur = cur+1
            del solver




            print('+++++++   Picard   +++++++')
            solver = Solver(local_epochs = m_, eta0=.1, C=0, algo_='Picard')
            solver.fit(self.A, self.k, W0=W0, NMF_=3)
            start_time = time.time()
            err = solver.train( maxIter=self.maxiter,warmup=False, damping=1)
            print(cur, 'time used:   ' ,time.time() - start_time)
            print('error', err)
            print()
            l = min(len(err), self.maxiter+1)
            errMat[cur, 0:l] = err[0:l]
            picards.append(solver.funcValsPerPicard)
            cur = cur+1
            del solver





        plot_all( self.k, m_ ,xname= 'number of Picard', picards=picards, AAs=AAs, AAPs=AAPs, resAAs=resAAs)
        plot_all( self.k, m_ ,xname= 'number of Picard', picards=[picards[0]], AAs=[AAs[0]], AAPs=[AAPs[0]], resAAs=[resAAs[0]])
        #plot_all( self.k, m_ ,xname= 'global iteration $t$', picards=[errMat[3]], AAs=[errMat[1]], AAPs=[errMat[0]], resAAs=[errMat[2]] )
        plot_median( self.k, m_ ,xname= 'number of Picard', picards=picards, AAs=AAs, AAPs=AAPs, resAAs=resAAs)
        plot_mean( self.k, m_ ,xname= 'number of Picard', picards=picards, AAs=AAs, AAPs=AAPs, resAAs=resAAs)



        return errMat





import matplotlib.pyplot as plt

# Function to calculate median and interquartile range
def compute_median(data_list):

    max_length = max(len(seq) for seq in data_list)
    padded_data = np.zeros((len(data_list), max_length))
    for i, seq in enumerate(data_list):
        padded_data[i, :len(seq)] = seq

    data_array = padded_data
    median = np.median(data_array, axis=0)
    q1 = np.percentile(data_array, 25, axis=0)
    q3 = np.percentile(data_array, 75, axis=0)
    return median, q1, q3

def compute_mean_and_std(data_list):

    max_length = max(len(seq) for seq in data_list)
    padded_data = np.zeros((len(data_list), max_length))
    for i, seq in enumerate(data_list):
        padded_data[i, :len(seq)] = seq

    data_array = padded_data
    mean = np.mean(data_array, axis=0)
    std = np.std(data_array, axis=0)
    return mean, std

def plot_median( k, m, xname='', damping = 1, picards=None, AAs=None, AAPs=None, resAAs=None):


    if picards is not None:
      median_AA, q1_AA, q3_AA = compute_median(picards)
      plt.semilogy(median_AA, label='Picard', color='tab:orange', linestyle=':', linewidth=1.5)
      plt.fill_between(range(len(median_AA)), q1_AA, q3_AA, color='tab:orange', alpha=0.2)

    if AAs is not None:
      median_AA, q1_AA, q3_AA = compute_median(AAs)
      plt.semilogy(median_AA, label='Picard', color='tab:brown', linestyle=':', linewidth=1.5)
      plt.fill_between(range(len(median_AA)), q1_AA, q3_AA, color='tab:brown', alpha=0.2)

    if resAAs is not None:
      median_AA, q1_AA, q3_AA = compute_median(resAAs)
      plt.semilogy(median_AA, label='Picard', color='tab:green', linestyle=':', linewidth=1.5)
      plt.fill_between(range(len(median_AA)), q1_AA, q3_AA, color='tab:green', alpha=0.2)

    if AAPs is not None:
      median_AA, q1_AA, q3_AA = compute_median(AAPs)
      plt.semilogy(median_AA, label='Picard', color='tab:red', linestyle=':', linewidth=1.5)
      plt.fill_between(range(len(median_AA)), q1_AA, q3_AA, color='tab:red', alpha=0.2)



    plt.legend()
    plt.title(f'NMF with r={k}, m={m}, median and interquartile range (IQR)')
    plt.xlabel(xname)  # Updated x-axis label
    #plt.ylabel('$||w^t - w^*||$')  # Updated y-axis label
    plt.tight_layout()

    plt.savefig(f'2_NMF with k={k}, m={m}'+xname+'.png')
    plt.show()

def plot_mean( k, m, xname='', damping = 1, picards=None, AAs=None, AAPs=None, resAAs=None):


    if picards is not None:
      mean, std = compute_mean_and_std(picards)
      plt.semilogy(mean, label='Picard', color='tab:orange', linestyle='-', linewidth=2.5)
      plt.fill_between(range(len(mean)), mean - std, mean + std, color='tab:orange', alpha=0.2)

    if AAs is not None:
      mean, std = compute_mean_and_std(AAs)
      plt.semilogy(mean, label='AA', color='tab:brown', linestyle='-', linewidth=2.5)
      plt.fill_between(range(len(mean)), mean - std, mean + std, color='tab:brown', alpha=0.2)

    if resAAs is not None:
      mean, std = compute_mean_and_std(resAAs)
      plt.semilogy(mean, label='resAA', color='tab:green', linestyle='-', linewidth=2.5)
      plt.fill_between(range(len(mean)), mean - std, mean + std, color='tab:green', alpha=0.2)

    if AAPs is not None:
      mean, std = compute_mean_and_std(AAPs)
      plt.semilogy(mean, label='AAP', color='tab:red', linestyle='-', linewidth=2.5)
      plt.fill_between(range(len(mean)), mean - std, mean + std, color='tab:red', alpha=0.2)



    plt.legend()
    plt.title(f'NMF with r={k}, m={m}, mean+-std')
    plt.xlabel(xname)  # Updated x-axis label
    #plt.ylabel('$||w^t - w^*||$')  # Updated y-axis label
    plt.tight_layout()

    plt.savefig(f'2_NMF with k={k}, m={m}'+xname+'.png')
    plt.show()

def plot_median( k, m, xname='', damping = 1, picards=None, AAs=None, AAPs=None, resAAs=None):


    if picards is not None:
      median_AA, q1_AA, q3_AA = compute_median(picards)
      plt.semilogy(median_AA, label='Picard', color='tab:orange', linestyle='-', linewidth=2.2)
      plt.fill_between(range(len(median_AA)), q1_AA, q3_AA, color='tab:orange', alpha=0.3)

    if AAs is not None:
      median_AA, q1_AA, q3_AA = compute_median(AAs)
      plt.semilogy(median_AA, label='AA(5)', color='tab:blue', linestyle='-', linewidth=2.2)
      plt.fill_between(range(len(median_AA)), q1_AA, q3_AA, color='tab:blue', alpha=0.1)

    if resAAs is not None:
      median_AA, q1_AA, q3_AA = compute_median(resAAs)
      plt.semilogy(median_AA, label='resAA(5)', color='tab:green', linestyle='-', linewidth=2.2)
      plt.fill_between(range(len(median_AA)), q1_AA, q3_AA, color='tab:green', alpha=0.3)

    if AAPs is not None:
      median_AA, q1_AA, q3_AA = compute_median(AAPs)
      plt.semilogy(median_AA, label='AAP(5)', color='tab:red', linestyle='-', linewidth=2.2)
      plt.fill_between(range(len(median_AA)), q1_AA, q3_AA, color='tab:red', alpha=0.3)



    plt.legend(loc='lower left')
    #plt.title(f'NMF with r={k}, m={m}, median and interquartile range (IQR)')
    plt.xlabel(xname, fontsize=12)
    plt.ylabel('$\|A - W_tH_t\|$ / $\| A\|$', fontsize=12)
    plt.tight_layout()

    plt.savefig(f'2_NMF with k={k}, m={m}'+xname+'.png')
    plt.show()

def plot_all( k, m, xname='', damping = 1, picards=None, AAs=None, AAPs=None, resAAs=None):


    if picards is not None:
      for i, picard in enumerate(picards):
        plt.semilogy(picard, label='Picard' if i == 0 else "", color='tab:orange', linestyle=':', linewidth=2.2)
    if AAs is not None:
      for i, AA in enumerate(AAs):
        plt.semilogy(AA, label='AA(5)'  if i == 0 else "", color='tab:blue', linestyle='--', linewidth=2.2)
    if resAAs is not None:
      for i, resAA in enumerate(resAAs):
        plt.semilogy(resAA, label='resAA(5)'  if i == 0 else "", color='tab:green', linestyle='-.', linewidth=2.2)#, marker='d', markevery=0.1, )
    if AAPs is not None:
      for i, AAP in enumerate(AAPs):
        plt.semilogy(AAP, label='AAP(5)' if i == 0 else "", color='tab:red', linestyle='-', marker='o',markevery=(5,6), linewidth=2.2, markersize=5)

    plt.legend(loc='lower left')
    #plt.title(f'NMF with k={k}, m={m}')
    plt.xlabel(xname, fontsize=12)
    plt.ylabel('$\|A - W_tH_t\|$ / $\| A\|$', fontsize=12)
    plt.tight_layout()

    plt.show()






if __name__ == '__main__':
    MaxIter = 20
    m_ = 5
    Repeat = 15
    #Gamma = .1
    IsSearch = False
    #Dtype_ = np.longdouble

    k = 4
    filename = f'/content/drive/MyDrive/Colab Notebooks/AA/github_main/AAP/Resource/nmf300_k{k}.npz'
    npzfile = numpy.load(filename)
    A = npzfile['A']
    print(A.shape, npzfile['W'].shape, npzfile['H'].shape)

    assert npzfile['W'].shape[1]==k, 'k does not match the ground truth'



    demo = Demo(MaxIter, Repeat, m_, None)
    demo.fit(A, k)
    errMat = demo.testConvergence(isSearch=IsSearch)


```
